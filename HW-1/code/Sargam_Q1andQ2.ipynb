```
import numpy as np
import matplotlib.pyplot as plt
import random
```

```
def EpsilonGreedy(variance,NoOfTimeSteps,NoOfRuns,NoOfArms,EpsilonVariesWithTime,GivenEpsilon=0):
  ExpectedReward=[0]*(NoOfTimeSteps)
  AverageAbsoluteError=[0]*(NoOfTimeSteps)
  OptimalArmChosen=[0]*(NoOfTimeSteps)
  for j in range(NoOfRuns):
    ActualQ=[]
    for i in range(NoOfArms):
      ActualQ.append(np.random.normal(0,1))
    OptimalArm=np.argmax(ActualQ)
    EstimatedQ=[0]*NoOfArms
    NoOfTimesChosen=[0]*NoOfArms
    for i in range(NoOfTimeSteps):
      x=random.random()
      if EpsilonVariesWithTime:
        epsilon=1/(i+1)
      else:
        epsilon=GivenEpsilon
      if x>1-epsilon:
        ChosenArm=int((random.random()*10))
      else:
        ChosenArm=np.argmax(EstimatedQ)
      if ChosenArm==OptimalArm:
        OptimalArmChosen[i]+=1
      CurrentReward=np.random.normal(ActualQ[ChosenArm],variance**(1/2))
      error=abs(CurrentReward-EstimatedQ[ChosenArm])
      NoOfTimesChosen[ChosenArm]+=1
      EstimatedQ[ChosenArm]+=(CurrentReward-EstimatedQ[ChosenArm])/NoOfTimesChosen[ChosenArm]
      AverageAbsoluteError[i]+=error
      ExpectedReward[i]+=CurrentReward
  for i in range(NoOfTimeSteps):
    ExpectedReward[i]/=NoOfRuns
    AverageAbsoluteError[i]/=NoOfRuns
    OptimalArmChosen[i]*=(100/NoOfRuns)
  return ExpectedReward,AverageAbsoluteError,OptimalArmChosen
```

## Question 1

```
ExpR1,AvgAbsErr1,OptArmPercent1=EpsilonGreedy(1,1000,2000,10,False,0) #epsilon=0
ExpR2,AvgAbsErr2,OptArmPercent2=EpsilonGreedy(1,1000,2000,10,False,0.01) #epsilon=0.01
ExpR3,AvgAbsErr3,OptArmPercent3=EpsilonGreedy(1,1000,2000,10,False,0.1) #epsilon=0.1
ExpR4,AvgAbsErr4,OptArmPercent4=EpsilonGreedy(1,1000,2000,10,True) #epsilon varies with time
```

Plotting Average Rewards


```
plt.plot(ExpR1)
plt.plot(ExpR2)
plt.plot(ExpR3)
plt.plot(ExpR4)
plt.legend(["Epsilon=0","Epsilon=0.01","Epsilon=0.1","Epsilon=1/t"])
plt.xlabel("Timesteps")
plt.ylabel("Average Reward")
plt.show()
```

Plotting Average Absolute Error

```
plt.plot(AvgAbsErr1)
plt.plot(AvgAbsErr2)
plt.plot(AvgAbsErr3)
plt.plot(AvgAbsErr4)
plt.legend(["Epsilon=0","Epsilon=0.01","Epsilon=0.1","Epsilon=1/t"])
plt.xlabel("Timesteps")
plt.ylabel("Average Absolute Error")
plt.show()
```

Plotting Optimal Action %

```
plt.plot(OptArmPercent1)
plt.plot(OptArmPercent2)
plt.plot(OptArmPercent3)
plt.plot(OptArmPercent4)
plt.legend(["Epsilon=0","Epsilon=0.01","Epsilon=0.1","Epsilon=1/t"])
plt.xlabel("Timesteps")
plt.ylabel("Optimal Action %")
plt.show()
```
